```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
sns.lmplot(data=data, x='Annual Income (k$)',y='Spending Score (1-100)', fit_reg=False)
```

```python
x = data.iloc[:,3:5].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=10, init='k-means++',random_state=10)
y_pred = kmeans.fit_predict(x)
```

```python
y_pred
```

```python
data
4 - LSLSS
2 - LSHSS
3 - HSHSS
0 - HSLSS
1 - ASASS
```

```python
output = pd.concat([data, pd.DataFrame(y_pred)],axis=1)
```

```python
output.rename(columns={0:'Cluster'},inplace=True)
```

```python
output['Cluster']=output.Cluster.map({4:'LSLSS',1:'ASASS',2:'LSHSS',3:'HSHSS',0:'HSLSS'})
```

```python
output.to_csv('Output_today.csv',index=False)
```
