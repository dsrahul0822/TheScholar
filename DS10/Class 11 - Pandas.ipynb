**%Contribution calculations**

```python
tips.groupby('gender').agg({'total_bill':np.mean})
```

```python
avg_sales = tips.groupby('gender').total_bill.transform('mean')
```

```python
avg_sales
```

```python
tips['avg_sales'] = avg_sales
```

```python
tips['diff'] = tips.total_bill-tips.avg_sales
```

```python
tips['cust_seg'] = tips['diff']>0
```

```python
tips['cust_seg'] = tips.cust_seg.map({False:'Below Average', True:'Above Average'})
```

```python
tips['gensmk'] = tips.gender + ',' + tips.smoker
```

```python
tips
```

```python
tips.groupby(['gender','cust_seg']).agg({'total_bill':np.mean})
```

```python
tips.groupby(['cust_seg','day']).total_bill.mean().unstack()
```

```python
tips.pivot_table(index='cust_seg', columns='day', values ='total_bill', aggfunc='mean')
```
