# Data Visualization

**Data Visualization gives us the faster way of understanding the data**


**Exploratory Data Analysis (EDA)**


    1) Univariate Analysis - Where we are taking only one variable and analysing that, for example using the histogram we can create the data distribution age of customers here we can use the box plot. For categorial variable for gender we can use the count plots which can help us to understand how many customers are male and female
    
    2) Bivariate Analysis - Where we are taking only two variables and analysing that. It can be numerical and numerical (scatter plot) For example the relationship between house price and the area of the house. It can be categorical and numerical Gender and the sales. 
    
    3) Multivariate Analysis - Where we are taking more than two variables, it can be all numericals, it can be mix of categoricals and numericals.

```python
dataset_loan = pd.read_csv('Python_Credit_Risk_XTrain.csv')
```

```python
dataset_loan
```

```python
dataset_loan.isnull().sum()
```

```python
for group, data in dataset_loan.groupby('Property_Area'):
    print(group,data['ApplicantIncome'].mean())
```

```python
plt.figure(figsize=(10,3))
for group, data in dataset_loan.groupby('Property_Area'):
    plt.bar(group,data['ApplicantIncome'].mean())
    plt.xlabel('Property Area')
    plt.ylabel('Avg. Applicant Income')
    plt.title('Property Area vs. Avg. Applicant Income')
```

```python
plt.figure(figsize=(10,3))
for group, data in dataset_loan.groupby('Property_Area'):
    plt.barh(group,data['ApplicantIncome'].mean())
    plt.xlabel('Property Area')
    plt.ylabel('Avg. Applicant Income')
    plt.title('Property Area vs. Avg. Applicant Income')
```

```python
prop_area = dict()
for i in set(dataset_loan['Property_Area']):
    prop_area[i] = dataset_loan[dataset_loan['Property_Area']==i].Property_Area.count()
```

```python
prop_area
```

```python
plt.figure(figsize=(5,3))
plt.bar(prop_area.keys(), prop_area.values())
plt.xlabel('Property Area')
plt.ylabel('Count of Property Area')
plt.title('Property Area Count')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.pie(prop_area.values(), labels=prop_area.keys())
plt.title('Property Area Count')
plt.show()
```

```python
plt.hist(dataset_loan.ApplicantIncome)
plt.show()
```

```python
plt.scatter(dataset_loan['ApplicantIncome'], dataset_loan['CoapplicantIncome'])
plt.xlabel('Applicant Income')
plt.ylabel('Coapplicant Income')
plt.xlim((0,20000))
plt.ylim((0,10000))
plt.show()
```
